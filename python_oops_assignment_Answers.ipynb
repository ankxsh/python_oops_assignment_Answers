```python
"""
Python OOPs Assignment - Complete Solutions
All theory questions answered and practical coding questions implemented
Ready to run on Google Colab
"""

print("="*80)
print("PYTHON OOPs ASSIGNMENT - THEORY QUESTIONS")
print("="*80)

# Theory Questions with Answers

theory_questions = {
    "Q1": {
        "question": "What is Object-Oriented Programming (OOP)?",
        "answer": """Object-Oriented Programming is a programming paradigm that organizes code
around objects and classes. It focuses on creating reusable code by modeling real-world entities
as objects with attributes (data) and methods (behavior). OOP is based on four main principles:
Encapsulation, Inheritance, Polymorphism, and Abstraction."""
    },

    "Q2": {
        "question": "What is a class in OOP?",
        "answer": """A class is a blueprint or template for creating objects. It defines the
attributes (data) and methods (functions) that objects of that class will have. Classes
encapsulate data and behavior together in a single unit."""
    },

    "Q3": {
        "question": "What is an object in OOP?",
        "answer": """An object is an instance of a class. It is a concrete entity created from
a class blueprint that has its own state (attribute values) and can perform actions through
methods. Multiple objects can be created from the same class."""
    },

    "Q4": {
        "question": "What is the difference between abstraction and encapsulation?",
        "answer": """Abstraction focuses on hiding complexity and showing only essential features
to the user. It deals with what an object does. Encapsulation is about bundling data and methods
together and restricting direct access to some components. It deals with how the object protects
its internal state."""
    },

    "Q5": {
        "question": "What are dunder methods in Python?",
        "answer": """Dunder methods (Double UNDERscore methods) are special methods in Python that
have double underscores before and after their names (e.g., __init__, __str__, __add__). They
are also called magic methods and allow us to define how objects of a class behave with built-in
operations like initialization, string representation, arithmetic operations, etc."""
    },

    "Q6": {
        "question": "Explain the concept of inheritance in OOP",
        "answer": """Inheritance is a mechanism where a new class (child/derived class) inherits
attributes and methods from an existing class (parent/base class). It promotes code reusability
and establishes a relationship between classes. The child class can override parent methods and
add its own attributes and methods."""
    },

    "Q7": {
        "question": "What is polymorphism in OOP?",
        "answer": """Polymorphism means 'many forms'. It allows objects of different classes to be
treated as objects of a common parent class. The same method name can behave differently based on
the object calling it. It can be achieved through method overriding and method overloading."""
    },

    "Q8": {
        "question": "How is encapsulation achieved in Python?",
        "answer": """Encapsulation in Python is achieved by using private and protected attributes.
Private attributes are prefixed with double underscore (__) and cannot be accessed directly from
outside the class. Protected attributes are prefixed with single underscore (_). Python also uses
property decorators to control access to attributes."""
    },

    "Q9": {
        "question": "What is a constructor in Python?",
        "answer": """A constructor is a special method __init__() that is automatically called when
an object is created. It initializes the object's attributes and sets up the initial state.
Python also has __new__() which is called before __init__() to create the object."""
    },

    "Q10": {
        "question": "What are class and static methods in Python?",
        "answer": """Class methods are defined using @classmethod decorator and receive the class
as the first parameter (cls). They can access and modify class state. Static methods are defined
using @staticmethod decorator and don't receive the class or instance as first parameter. They
behave like regular functions but belong to the class namespace."""
    },

    "Q11": {
        "question": "What is method overloading in Python?",
        "answer": """Method overloading allows multiple methods with the same name but different
parameters. Python doesn't support traditional method overloading like Java. Instead, we can
achieve similar functionality using default arguments, variable-length arguments (*args, **kwargs),
or by checking argument types within the method."""
    },

    "Q12": {
        "question": "What is method overriding in OOP?",
        "answer": """Method overriding occurs when a child class provides a specific implementation
of a method that is already defined in its parent class. The child class method has the same name,
parameters, and return type as the parent class method but different implementation."""
    },

    "Q13": {
        "question": "What is a property decorator in Python?",
        "answer": """The @property decorator allows us to define methods that can be accessed like
attributes. It provides a way to implement getters, setters, and deleters for class attributes,
allowing controlled access and validation while maintaining a simple attribute-like syntax."""
    },

    "Q14": {
        "question": "Why is polymorphism important in OOP?",
        "answer": """Polymorphism is important because it promotes code flexibility and reusability.
It allows writing generic code that can work with objects of different classes, making the code
more maintainable and extensible. It enables us to write one interface and have multiple
implementations."""
    },

    "Q15": {
        "question": "What is an abstract class in Python?",
        "answer": """An abstract class is a class that cannot be instantiated and is meant to be
subclassed. It contains one or more abstract methods (methods without implementation) that must
be implemented by child classes. In Python, abstract classes are created using the ABC module
and @abstractmethod decorator."""
    },

    "Q16": {
        "question": "What are the advantages of OOP?",
        "answer": """Main advantages include: 1) Code reusability through inheritance, 2) Data
security through encapsulation, 3) Flexibility through polymorphism, 4) Easy maintenance and
modification, 5) Better code organization and modularity, 6) Real-world modeling capability,
7) Reduced complexity through abstraction."""
    },

    "Q17": {
        "question": "What is the difference between a class variable and an instance variable?",
        "answer": """Class variables are shared among all instances of a class and are defined
outside any method. Instance variables are unique to each instance and are defined inside __init__
method using self. Class variables are accessed using the class name or instance, while instance
variables are accessed using self within methods."""
    },

    "Q18": {
        "question": "What is multiple inheritance in Python?",
        "answer": """Multiple inheritance is when a class inherits from more than one parent class.
Python supports multiple inheritance, and the child class inherits attributes and methods from
all parent classes. Method Resolution Order (MRO) determines which parent's method is called
when there are conflicts."""
    },

    "Q19": {
        "question": "Explain the purpose of '__str__' and '__repr__' methods in Python",
        "answer": """__str__ returns a human-readable string representation of the object, used by
print() and str(). __repr__ returns an unambiguous string representation that ideally could be
used to recreate the object, used by repr() and in interactive mode. __str__ is for end users,
__repr__ is for developers."""
    },

    "Q20": {
        "question": "What is the significance of the 'super()' function in Python?",
        "answer": """super() is used to call methods from a parent class in the child class. It
provides a way to access inherited methods that have been overridden. It's especially useful in
multiple inheritance to follow the Method Resolution Order (MRO) correctly."""
    },

    "Q21": {
        "question": "What is the significance of the __del__ method in Python?",
        "answer": """__del__ is a destructor method called when an object is about to be destroyed
or garbage collected. It's used to perform cleanup operations like closing files or network
connections. However, it's not guaranteed when it will be called, so it's not always reliable."""
    },

    "Q22": {
        "question": "What is the difference between @staticmethod and @classmethod in Python?",
        "answer": """@staticmethod doesn't receive any implicit first argument (no self or cls).
It can't access or modify class or instance state. @classmethod receives the class as the first
argument (cls) and can access or modify class state. Class methods are often used as factory
methods."""
    },

    "Q23": {
        "question": "How does polymorphism work in Python with inheritance?",
        "answer": """Polymorphism with inheritance allows child classes to override parent class
methods with their own implementations. When a method is called on an object, Python uses the
object's actual class (not the reference type) to determine which method to execute, enabling
different behaviors for the same method call."""
    },

    "Q24": {
        "question": "What is method chaining in Python OOP?",
        "answer": """Method chaining is a technique where multiple methods are called on the same
object in a single statement. Each method returns 'self' (the object itself), allowing the next
method to be called on the returned object. This creates a fluent interface."""
    },

    "Q25": {
        "question": "What is the purpose of the __call__ method in Python?",
        "answer": """The __call__ method allows an instance of a class to be called as a function.
When you use parentheses on an object (obj()), Python calls the __call__ method. This is useful
for creating callable objects, decorators, or implementing function-like behavior in classes."""
    }
}

# Print all theory questions and answers
for q_num, q_data in theory_questions.items():
    print(f"\n{q_num}. {q_data['question']}")
    print(f"Answer: {q_data['answer']}")
    print("-"*80)

print("\n" + "="*80)
print("PRACTICAL CODING QUESTIONS - IMPLEMENTATIONS")
print("="*80)

# ==================== PRACTICAL QUESTION 1 ====================
print("\n" + "="*80)
print("Q1: Parent class Animal with speak() method and child class Dog")
print("="*80)

class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Bark!")

# Test
animal = Animal()
dog = Dog()
print("Animal speak:")
animal.speak()
print("Dog speak:")
dog.speak()

# ==================== PRACTICAL QUESTION 2 ====================
print("\n" + "="*80)
print("Q2: Abstract class Shape with Circle and Rectangle")
print("="*80)

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Test
circle = Circle(5)
rectangle = Rectangle(4, 6)
print(f"Circle area (radius=5): {circle.area():.2f}")
print(f"Rectangle area (4x6): {rectangle.area()}")

# ==================== PRACTICAL QUESTION 3 ====================
print("\n" + "="*80)
print("Q3: Multi-level inheritance - Vehicle, Car, ElectricCar")
print("="*80)

class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type

    def display_type(self):
        print(f"Vehicle type: {self.type}")

class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

    def display_info(self):
        print(f"Car - Type: {self.type}, Brand: {self.brand}")

class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery):
        super().__init__(vehicle_type, brand)
        self.battery = battery

    def display_full_info(self):
        print(f"Electric Car - Type: {self.type}, Brand: {self.brand}, Battery: {self.battery}")

# Test
electric_car = ElectricCar("Sedan", "Tesla", "100 kWh")
electric_car.display_full_info()

# ==================== PRACTICAL QUESTION 4 ====================
print("\n" + "="*80)
print("Q4: Polymorphism - Bird, Sparrow, Penguin")
print("="*80)

class Bird:
    def fly(self):
        print("This bird can fly")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high in the sky")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly, but it can swim")

# Test
birds = [Sparrow(), Penguin()]
for bird in birds:
    bird.fly()

# ==================== PRACTICAL QUESTION 5 ====================
print("\n" + "="*80)
print("Q5: Encapsulation - BankAccount class")
print("="*80)

class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid withdrawal amount or insufficient balance")

    def check_balance(self):
        print(f"Current balance: ${self.__balance}")
        return self.__balance

# Test
account = BankAccount(1000)
account.check_balance()
account.deposit(500)
account.withdraw(300)
account.check_balance()

# ==================== PRACTICAL QUESTION 6 ====================
print("\n" + "="*80)
print("Q6: Runtime Polymorphism - Instrument, Guitar, Piano")
print("="*80)

class Instrument:
    def play(self):
        print("Playing an instrument")

class Guitar(Instrument):
    def play(self):
        print("Strumming the guitar strings")

class Piano(Instrument):
    def play(self):
        print("Pressing the piano keys")

# Test
instruments = [Guitar(), Piano()]
for instrument in instruments:
    instrument.play()

# ==================== PRACTICAL QUESTION 7 ====================
print("\n" + "="*80)
print("Q7: Class method and static method - MathOperations")
print("="*80)

class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

# Test
print(f"Addition (5 + 3): {MathOperations.add_numbers(5, 3)}")
print(f"Subtraction (10 - 4): {MathOperations.subtract_numbers(10, 4)}")

# ==================== PRACTICAL QUESTION 8 ====================
print("\n" + "="*80)
print("Q8: Class method to count persons created")
print("="*80)

class Person:
    person_count = 0

    def __init__(self, name):
        self.name = name
        Person.person_count += 1

    @classmethod
    def get_person_count(cls):
        return cls.person_count

# Test
p1 = Person("Alice")
p2 = Person("Bob")
p3 = Person("Charlie")
print(f"Total persons created: {Person.get_person_count()}")

# ==================== PRACTICAL QUESTION 9 ====================
print("\n" + "="*80)
print("Q9: Fraction class with __str__ override")
print("="*80)

class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero")
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

# Test
frac = Fraction(3, 4)
print(f"Fraction: {frac}")

# ==================== PRACTICAL QUESTION 10 ====================
print("\n" + "="*80)
print("Q10: Operator overloading - Vector class")
print("="*80)

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

# Test
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2
print(f"v1: {v1}")
print(f"v2: {v2}")
print(f"v1 + v2: {v3}")

# ==================== PRACTICAL QUESTION 11 ====================
print("\n" + "="*80)
print("Q11: Person class with greet method")
print("="*80)

class Person2:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Test
person = Person2("John", 25)
person.greet()

# ==================== PRACTICAL QUESTION 12 ====================
print("\n" + "="*80)
print("Q12: Student class with average_grade method")
print("="*80)

class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

# Test
student = Student("Emma", [85, 90, 78, 92, 88])
print(f"Student: {student.name}")
print(f"Grades: {student.grades}")
print(f"Average grade: {student.average_grade():.2f}")

# ==================== PRACTICAL QUESTION 13 ====================
print("\n" + "="*80)
print("Q13: Rectangle class with set_dimensions and area methods")
print("="*80)

class Rectangle2:
    def __init__(self):
        self.length = 0
        self.width = 0

    def set_dimensions(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Test
rect = Rectangle2()
rect.set_dimensions(5, 8)
print(f"Rectangle dimensions: {rect.length} x {rect.width}")
print(f"Area: {rect.area()}")

# ==================== PRACTICAL QUESTION 14 ====================
print("\n" + "="*80)
print("Q14: Employee and Manager classes with salary calculation")
print("="*80)

class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()
        return base_salary + self.bonus

# Test
emp = Employee("Ankush", 160, 20)
mgr = Manager("Archita", 160, 30, 1000)
print(f"Employee {emp.name} salary: ${emp.calculate_salary()}")
print(f"Manager {mgr.name} salary: ${mgr.calculate_salary()}")

# ==================== PRACTICAL QUESTION 15 ====================
print("\n" + "="*80)
print("Q15: Product class with total_price method")
print("="*80)

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity

# Test
product = Product("Laptop", 800, 3)
print(f"Product: {product.name}")
print(f"Price per unit: ${product.price}")
print(f"Quantity: {product.quantity}")
print(f"Total price: ${product.total_price()}")

# ==================== PRACTICAL QUESTION 16 ====================
print("\n" + "="*80)
print("Q16: Abstract Animal class with Cow and Sheep")
print("="*80)

class Animal2(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal2):
    def sound(self):
        print("Moo!")

class Sheep(Animal2):
    def sound(self):
        print("Baa!")

# Test
animals = [Cow(), Sheep()]
for animal in animals:
    animal.sound()

# ==================== PRACTICAL QUESTION 17 ====================
print("\n" + "="*80)
print("Q17: Book class with get_book_info method")
print("="*80)

class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"'{self.title}' by {self.author}, published in {self.year_published}"

# Test
book = Book("Life", "Ankush", 2025)
print(book.get_book_info())

# ==================== PRACTICAL QUESTION 18 ====================
print("\n" + "="*80)
print("Q18: House and Mansion classes")
print("="*80)

class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    def display_info(self):
        print(f"Address: {self.address}, Price: ${self.price}")

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def display_info(self):
        print(f"Mansion at {self.address}, Price: ${self.price}, Rooms: {self.number_of_rooms}")

# Test
house = House("Sector 13", 250000)
mansion = Mansion("Dwarka", 2000000, 15)
house.display_info()
mansion.display_info()

print("\n" + "="*80)

```
